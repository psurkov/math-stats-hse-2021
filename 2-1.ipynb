```python
import numpy as np
from scipy.special import factorial
import matplotlib.pyplot as plt
```

```python
b = 10
lam = 5
size = 1000
```

```python
def uniform(k):
    mean = np.mean(np.random.uniform(0, b, size) ** k)
    return ((k + 1) * mean) ** (1/k)

def exp(k):
    mean = np.mean(np.random.exponential(lam, size) ** k)
    return (mean / factorial(k)) ** (1/k)
```

```python
def std(p, m):
    return np.mean((p - m) ** 2)
```

```python
stdu = []
stde = []
len = 50
x = np.arange(1, len)
for k in range(1, len):
    us = []
    es = []
    for i in range(1000):
        us.append(uniform(k))
        es.append(exp(k))
    stdu.append(std(np.array(us), b))
    stde.append(std(np.array(es), lam))
```

```python
plt.title("uniform")
plt.xlabel("k")
plt.ylabel("std")
plt.plot(x, stdu)
```

```python
plt.title("exp")
plt.xlabel("k")
plt.ylabel("std")
plt.plot(x, stde)
```
